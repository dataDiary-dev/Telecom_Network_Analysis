Handle Missing Values

```python
#check missing values
df.isnull().sum()
#fill missing vlaues
df['users_connected'] = df['users_connected'].fillna(0)
```

```python
df['download_speed'] = df['download_speed'].fillna(df['download_speed'].mean())
```

```python
df['upload_speed'] = df['upload_speed'].fillna(df['upload_speed'].mean())
```

```python
df['latency'] = df['latency'].fillna(df['latency'].mean())
```

```python
df['weather'] = df['weather'].fillna("Unknown")
```

Basic Filtering

```python
#high latency cases
high_latency = df[df['latency'] > 100]
high_latency.head()
```

```python
#low download speed
low_speed = df[df['download_speed'] < 10]
low_speed.head()
```

```python
#specific tower
tower_data = df[df['tower_id'].isin([1, 2, 3])]
tower_data.head()
```

Simple analysis

```python
#average speeds
df[['download_speed','upload_speed']].mean()

#users vs speed
df[['users_connected', 'download_speed']].corr()
```
